```
import spacy
from tqdm import tqdm
import os
```

```
path = "/content/drive/MyDrive/CS5260 Project/CrossNER/ner_data/"
```

```
# Setting up Spacy POS tagger
nlp = spacy.load("en_core_web_sm")
suffixes = nlp.Defaults.suffixes + (r'''\w+-\w+-\w+''',) + (r'''[0-9][a-zA-Z]''',) + (r'''[a-zA-Z][0-9][A-Z]''',) + (r'''\w+-\w+''',) + (r'''\w+-\w+-\w+-\w+''',)
suffixes = list(suffixes)
suffix_regex = spacy.util.compile_suffix_regex(suffixes)
nlp.tokenizer.suffix_search = suffix_regex.search


# Starting tagging
domain_list = ["ai", "conll2003", "literature", "music", "politics", "science"]
dic = {}
for domain in tqdm(domain_list):
  full_path= path + domain
  file_merge = []
  for file in tqdm(os.listdir(full_path)):
    # Read textfile
    f = open(full_path+"/"+file, "r")
    content = f.read()

    # Split text into sentences from the original file
    sentences = content.split("\n\n")

    # Create a list to store all the words' information (POS and entity tag) from this text file
    sentence_merge = []

    # Go through each sentence
    for element in sentences:
      # Extract word vectors ([word, entity])
      words = element.split("\n")

      # Create a list to store word texts, which is used to combine and generate a full sentence later.  
      words_list = []

      # Create a dic to store entity tags for each word
      word_entity_dic = {}

      # Go through each word vector
      for word_info in words:
        try:
          # Extract word text
          word = word_info.split("\t")[0]
          # Extract entity tags
          word_entity = word_info.split("\t")[1]

          words_list.append(word)
          word_entity_dic[word] = word_entity

        except:
          continue;
          print(word_info)
      # Merge into a full sentence
      sentence = " ".join(words_list)
      # Feed into spacy POS tagger
      doc = nlp(sentence)
      # Create a list to store word_text, POS and entity tags for each word
      words_pos = []

      for token in doc:
        # Only pronouns and nouns are left
        if (token.pos_ == 'PRON') | (token.pos_ == 'NOUN'):
          try:
            words_pos.append([token.text, token.pos_, word_entity_dic[token.text]])
          except:
            words_pos.append([token.text, token.pos_, "None"])
      sentence_merge.append(words_pos)
    file_merge.append(sentence_merge)
  dic[domain] = file_merge
```

```
# Reorganize the dictionary
for i in tqdm(domain_list):
  flatten_list = [item for sublist in dic[i] for subsublist in sublist for item in subsublist]
  sub_dict = {"NOUN":[], "PRON": []}
  for j in tqdm(flatten_list):
    if j[1] == "NOUN":
      sub_dict["NOUN"].append(j)
    else:
      sub_dict["PRON"].append(j)
  dic[i] = sub_dict
```

```
import pickle
with open('/content/drive/MyDrive/CS5260 Project/CrossNER_NOUN_PRON.pickle', 'wb') as handle:
    pickle.dump(dic, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
